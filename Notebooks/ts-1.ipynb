### RNN for time series modeling

Using RNN for modeling synthesized time series data

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```
# generating synthesized sine curve and time series data for modeing
amplitude = 10
freq = .5
time = np.arange(0,100) 
data = amplitude*np.sin(time*freq)
#plotting the data
plt.plot(time,data)
plt.xlabel ('time')
plt.ylabel("amplitude")
plt.title('time series data')
plt.grid(True)
plt.show()
```

```
#generating a time series dataframe
idx = pd.date_range('2019-1-1',periods=100,freq='D')
df = pd.DataFrame(data,index=idx,columns=['Amplitude'])
df.head()
```
